```python
import pandas as pd
```

```python
csv_file_1 = "201601-citibike-tripdata.csv"
citibike1_data_df = pd.read_csv(csv_file_1)
citibike1_data_df.head()
```

```python
# Define a function to map the value
def set_value(row_number, assigned_value): 
    return assigned_value[row_number]
```

```python
# Create the dictionary 
gender_dictionary = {1:'male',2:'female',0:'unknow'}
```

```python
# Add a new column named 'gender category'
citibike1_data_df['gender category']=citibike1_data_df['gender'].apply(set_value,args=(gender_dictionary,))
```

```python
# drop n/a
citibike1_data_df = citibike1_data_df.dropna(how='any').reset_index(drop=True)
# Change Data Types
citibike1_data_df['birth year'] = citibike1_data_df['birth year'].astype(int)
citibike1_data_df['starttime'] = pd.to_datetime(citibike1_data_df['starttime'])
citibike1_data_df['stoptime'] = pd.to_datetime(citibike1_data_df['stoptime'])
# Add Age Column 
citibike1_data_df['Age'] = 2017 - citibike1_data_df['birth year']
# exclude ages > 90 years
citibike1_data_df = citibike1_data_df[citibike1_data_df['Age'] < 90]
    
citibike1_data_df.head()
```
